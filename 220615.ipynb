```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
url = 'https://www.kweather.co.kr/forecast/forecast_living_jisu.html'
response = requests.get(url)
```

```python
status_code = response.status_code
print(status_code)
```

```python
text = response.text
print(text)
```

```python
soup = BeautifulSoup(text, 'html.parser')
```

```python
print(soup)
```

```python
jisu_navi_content = soup.find(id='jisu_navi_content')
```

```python
print(jisu_navi_content)
```

```python
jisu_navi_content_li = jisu_navi_content.find_all('span')
```

```python
print(jisu_navi_content_li)
```

```python
urls = []
for i in jisu_navi_content_li:
    #print(i)
    data = i.a['href']
    #print(data)
    info = re.search('/data/JISU/.+.xml', data)
    #print(info.group())
    urls.append(info.group())
```

```python
for j in urls:
    print(j)
```

```python
url = 'http://www.kweather.co.kr'+urls[0]
response = requests.get(url)
status_code = response.status_code
print(status_code)
```

```python
print(response.encoding)
```
