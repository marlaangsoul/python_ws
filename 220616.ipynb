```python
import folium
```

```python
map_thej = folium.Map(location=(41.40338, 2.17403), zoom_start=30, tiles='Stamen Terrain')
map_thej
```

```python
map_thej = folium.Map(location=(41.40338, 2.17403), zoom_start=17)
folium.Marker((41.40338, 2.17403), popup="바티칸").add_to(map_thej)
map_thej
```

```python
map_thej = folium.Map(location=(41.40600, 2.17700), zoom_start=17)
folium.Marker((41.40600, 2.17700), popup="바티칸 정원", icon=folium.Icon(color='red', icon='info-sign')).add_to(map_thej)
map_thej
```

```python
map_thej6 = folium.Map(location=(41.40600, 2.17700), zoom_start=17)
folium.Marker((41.40600, 2.17700), popup="더조은부평", icon=folium.Icon(color='red', icon='info-sign')).add_to(map_thej6)
folium.Circle((41.40338, 2.17403), popup="더조은부평", radius=100, color='#3311cc', fill_color='#3311cc').add_to(map_thej6)
map_thej6
```
